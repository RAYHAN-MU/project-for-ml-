```python
#linear regresion
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

x = [10,11,15,20,25]
y = [20,22,20,21,27]

slope, intercept, r, p, std_err = stats.linregress(x, y) # This line calculates the slope and intercept using linear regression.
def myfun(x):
  return slope *x + intercept

yvalue=list(map(myfun,x))

print(yvalue)
print(y)
print(x)
plt.scatter(x,y)
plt.scatter(x,yvalue,c="red")
plt.plot(x,yvalue,c="yellow")
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Example dataset
data = {
    'Feature1': [10, 20, 30, 40, 50],
    'Feature2': [100, 200, 300, 400, 500]
}
df = pd.DataFrame(data)

# Normalization (Min-Max Scaling)
min_max_scaler = MinMaxScaler()
normalized_data = min_max_scaler.fit_transform(df)
normalized_df = pd.DataFrame(normalized_data, columns=df.columns)
print("Normalized Data:\n", normalized_df)

# Standardization (Z-Score Scaling)
standard_scaler = StandardScaler()
standardized_data = standard_scaler.fit_transform(df)
standardized_df = pd.DataFrame(standardized_data, columns=df.columns)
print("\nStandardized Data:\n", standardized_df)
```

```python
import numpy
import matplotlib.pyplot as plt
np.random.seed(2)
# Generate random data
x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x
# Fit a polynomial model
model = numpy.poly1d(numpy.polyfit(x, y, 4))
print(model)
# Generate x values for predictions
x_value = numpy.linspace(min(x), max(x), 100)
# Predict y values using the model
predict = model(x_value)
plt.scatter(x,y)
# Plot the polynomial fit
plt.plot(x_value, predict, color="red")
# Show the plot
plt.show()
```
