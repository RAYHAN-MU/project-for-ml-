```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report

```

```python
# Load dataset
df=pd.read_csv('/boot/dataset/marital status.csv')
print("your total dataset null value is ",df.isnull().sum())
```

```python
# Exploratory Data Analysis (EDA)
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
# Handling missing values
mode=df['status'].mode()[0]
df['status']=df.status.fillna(mode)
df
```

```python
# Splitting into training and testing sets
x=df['age'].values.reshape(-1,1)
y=df['status']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1,stratify=y)
```

```python
# Model Training
logisreg=LogisticRegression()
logisreg.fit(x_train,y_train)
y_pred=logisreg.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print("your accuracy is ",accuracy)
```

```python
# Predictions
sigmoidValue=logisreg.predict_proba(x_test)
print(y_pred)

print(sigmoidValue)
```

```python
#Model Evaluation
cm=confusion_matrix(y_test,y_pred)
print(cm)
```
